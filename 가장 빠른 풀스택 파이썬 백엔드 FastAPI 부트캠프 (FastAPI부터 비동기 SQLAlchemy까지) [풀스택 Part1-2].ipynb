```python
"""
타입 힌트
- 변수나 함수의 예상되는 데이터 타입을 명시적으로 표시하는 프로그래밍 기술
- FastAPI는 이를 활용하여 요청의 유효성을 검증하고 적절한 데이터 처리를 돕는다.

기본 타입 힌트 사용
- 경로나 쿼리 매개변수에 타입을 지정하여 자동으로 검증한다.
- int, float, str, bool 등이 가능하다.
---
from fastapi import FastAPI

app = FastAPI()

@app.get("/items/{item_id}")
def read_item(item_id: int):
    return {"item_id": item_id}

@app.get("/getdata/")
def read_item(data: str = "pfeelonoe"):
    return {"data": data}
---
- 타입 힌트와 맞지 않는 데이터를 입력할 경우 에러가 발생한다.
- 타입 힌트가 str인 곳에 숫자를 넣으면 str로 변환하여 입력된다.

고급 타입 힌트 사용
- typing 모듈의 List, Dict 등을 사용하여 복잡한 데이터 구조를 표현한다.
---
from fastapi import FastAPI, Query
from typing import List, Dict

app = FastAPI()

@app.get("/items/")
# List 타입 힌트는 Query와 한 몸으로 사용한다고 보면 된다.
# Data의 타입이 리스트이고 각 값은 숫자이어야 한다.
async def read_item(q: List[int] = Query([])):
    return {"q": q}

@app.post("/create-item/")
# 키는 str, 밸류는 int 타입이어야 한다.
async def create_item(item: Dict[str, int]):
    return item
---
- Query는 들어온 데이터를 유효성 검사하는 기능을 하며 ([])는 디폴트 값으로 리스트에
  아무런 값이 들어오지 않을 경우 []로 들어간다.
"""
```

```python
"""
HTTP 메서드
- 클라이언트가 서버에 특정 동작을 요청하는 방법을 정의하며 FastAPI에서는 이를 통해
  요청의 의도를 명확히하고 적절한 엔드포인트로 라우팅한다.
  
HTTP 메소드
1. GET
 - 서버로부터 정보 요청
 - 데이터 읽기 전용이며 서버 상태에 변경 없음
2. POST
 - 서버에 데이터 전송, 새 리소스 생성
 - 데이터를 서버에 제출하며 주로 요청 본문에 데이터 포함
3. PUT
 - 지정된 리소스의 전체 업데이트
 - 기존 리소스의 완전한 교체
4. DELETE
 - 지정된 리소스 삭제
 - 리소스 제거 지시, 성공 시 접근 불가
"""
```

```python
"""
2xx (성공)
- 200 OK: 성공적으로 처리
- 201 Created: 요청이 성공적으로 이루어져 새로운 리소스가 생성됨. 예를 들어 POST
  요청을 통해 새로운 엔트리가 데이터베이스에 추가되었을 때 사용
- 204 No Content: 요청은 성공적이지만 클라이언트에게 보내줄 콘텐츠는 없음

3xx (리다이렉션)
- 301 Moved Permanently: 요청한 리소스가 영구적으로 새 위치로 이동함 (URL 변경)
- 302 Found: 요청한 리소스가 일시적으로 다른 위치에 있음 (리다이렉션을 위해 자주 사용됨)

4xx (클라이언트 오류)
- 400 Bad Request: 서버가 요청을 이해할 수 없음. 잘못된 요청 구조나 파라미터 때문에
  발생
- 401 Unauthorized: 요청이 인증을 필요로 함
- 403 Forbidden: 서버가 요청을 이해했으나 권한이 없어 요청을 거부
- 404 Not Found: 서바가 요청한 리소스를 찾을 수 없음 (URL 오류 또는 존재하지 않는 페이지)
- 405 Method Not Allowed: 요청한 리소스에서는 지원하지 않는 HTTP 메소드 사용
- 422 Unprocessable Entity: 요청은 이해했으나 요청된 작업을 수행할 수 없음. 주로 요청 형식이
  올바르나 요청 내용이 유효하지 않을 떄 사용(데이터 유효성 검증 실패 등)
  
5xx (서버오류)
- 500 Internal Server Error: 서버 내부 오류로 인해 요청을 처리할 수 없음
- 503 Service Unvavailable: 서버가 일시적으로 요청을 처리할 수 없음. 주로 서버 과부하나
  유지관리로 인해 발생
  

"""
```

```python
"""
데이터베이스와 ORM
- FastAPI에서 데이터베이스 관리와 ORM(Object-Relational Mapping)의 사용은 웹 애플리케이션의
  중요한 부분으로 ORM은 데이터베이스의 테이블을 객체로 매핑하여 개발자가 SQL 쿼리 없이
  데이터베이스를 쉽게 조작할 수 있게 한다.

필요 라이브러리 설치
- pip install sqlalchemy
  pip install pymysql

데이터베이스 모델 정의
- SQLAlchemy를 사용하여 데이터베이스 모델을 정의하며 각 모델은 데이터베이스의 테이블에
  해당하고 모델의 인스턴스는 테이블의 레코드를 나타낸다.
---
from fastapi import FastAPI, Depends
from sqlalchemy.orm import Session
from sqlalchemy import Column, Integer, String, create_engine
from sqlalchemy.ext.declarative import declarative_base
from pydantic import BaseModel

# mysql을 pymysql 라이브러리로 사용
# 사용자 이름, 비밀번호, 서버주소, 데이터베이스 이름
DATABASE_URL = "mysql+pymysql://pfeelon:pfeelone@localhost/db_name"

# 해당 데이터베이스에 접근할 수 있는 객체 생성 (데이터베이스 엔진 생성)
engine = create_engine(DATABASE_URL)

# SQLAlchemy의 모델 기본 클래스를 선언하고 이 클래스를 상속받아 데이터베이스 테이블을 정의
# SQLAlchemy는 데이터베이스의 테이블도 객체로 관리
Base = declarative_base()

class User(Base):

    # users 테이블 정의
    __tablename__ = "users"
    
    # 각 컬럼 정의
    id = Column(Integer, primary_key=True, index=True) # 기본키로 설정하고 해당 컬럼 기반으로 인덱스를 만듦 (값을 받지 않아도 자동으로 할당됨)
    username = Column(String(50), unique=True, index=True) # 중복이 불가하며 인덱스를 만듦
    email = Column(String(120)) # 글자 길이를 120자로 제한
---
"""
```

```python
"""
데이터베이스 세션 생성 및 관리
- 우선 DB에 테이블이 생성되어 있어야 한다.
---
# Pydantic 모델을 정의하고 클라이언트로부터 받은 데이터의 유효성을 검사하는데 사용한다.
class UserCreate(BaseModel):
    username: str
    email: str

# 데이터베이스 세션을 생성하고 관리하는 의존성 함수를 정의한다.
def get_db(): # db 액세스하는 코드마다 내부적으로 호출되는 함수
    db = Session(bind=engine) # 데이터베이스 연결
    try:
        yield db
    finally:
        db.close()

# 데이터베이스 엔진을 사용하여 모델(User)을 기반으로 테이블을 생성한다.
Base.metadata.create_all(bind=engine)
---

FastAPI 애플리케이션에 모델 통합
---
app = FastAPI()

@app.get("/")
def read_root():
    return {"message": "Hello, World!"}
    
@app.post("/users/")
def create_user(user: UserCreate, db: Session = Depends(get_db)):
    new_user = User(username=user.username, email=user.email)
    db.add(new_user) # 데이터 삽입
    db.commit() # 데이터 커밋
    db.refresh(new_user) # 데이터베이스로부터 새 User 인스턴스의 최신 정보 가져옴
    return {"id": new_user.id, "username": new_user.username, "email": new_user.email}
---
"""
```

```python
"""
Column에서 사용할 수 있는 옵션
1. Data Types
 - Integer, String(최대 길이 명시), Boolean, FLoat, DateTime
2. Primary Key
3. Index
4. Unique
5. Nullable
6. Default Value
 - datetime.utcnow는 현재 시간을 나타낸다.
7. Auto Increment
 - 해당 필드의 값을 자동으로 증가시킨다.
8. Foreign Key
 - parent_id = Column(Integer, ForeignKey('parents.id')) # parents 테이블의 id 컬럼을 참조
9. Comments
 - 해당 필드의 설명
 
SQLAlchemy와 CRUD
- Session은 SQLAlchemy에서 데이터베이스와의 모든 상호작용을 관리하며 CRUD 작업을 위해
  세션을 사용한다. (데이터베이스와의 연결 관리)
- 세션 관리 시 Depends 또는 sessionmaker를 사용한다.

Depends
---
# Session 객체를 생성하고 요청 처리 후에 자동으로 정리하는 의존성 함수를 정의한다.
def get_db():
    db = Session(bind=engine)
    try:
        yield db
    finally:
        db.close()

Base.metadata.create_all(bind=engine)

app = FastAPI()

@app.get("/")
def read_root():
    return {"message": "Hello, World!"}
    
@app.post("/users/")
# 데이터베이스 접근을 위해 Depends를 매개변수로 추가한다.
def create_user(user: UserCreate, db: Session = Depends(get_db)):
    # User 모델 인스턴스를 생성하아ㅕ 데이터베이스 객체를 준비한다.
    new_user = User(username=user.username, email=user.email)
    # add() 함수로 인스턴스를 추가한다.
    db.add(new_user)
    db.commit()
    db.refresh(new_user)
    return {"id": new_user.id, "username": new_user.username, "email": new_user.email}
---

sessionmaker
- SessionLocal 팩토리는 sessionmaker를 이용해 데이터베이스 세션을 생성한다.
---
from sqlalchemy.orm import sessionmaker

# autocommit은 데이터베이스 작업 후에 자동으로 커밋할지의 여부를 결정하며 일반적으로
  트랜잭션(커밋 이전 변경사항) 관리를 위해 False로 설정한다.
# autoflush=False는 자동 플러시 비활성화 옵션이다. (변경사항이 DB 쿼리를 실행하기 전에
  자동으로 DB에 반영되어야 하는지의 여부를 결정)
SessionLocal = sessionmaker(autocommit=False, autoflush=False, bind=engine)

@app.post("/users/")
def create_user(user: UserCreate):
    # 데이터베이스 세션을 생성 (get_db 함수가 필요 없다.)
    db = SessionLocal()
    # User 인스턴스 생성 및 초기화
    db_user = User(username=user.username, email=user.email)
    db.add(new_user)
    db.commit()
    db.refresh(new_user)
    db.close() # get_db 함수가 없으므로 반드시 선언 필요
    return {"id": new_user.id, "username": new_user.username, "email": new_user.email}
---
"""
```

```python
"""
SQLAlchemy 비동기 문법
- SQLAlchemy 비동기 관련 기능이 동기 기능보다 최근에 개발 되었고 계속 개발 중이므로 관련 문법과 사용법이
  수시로 바뀌는 경향이 있다.

sqlalchemy.ext.asyncio
- 비동기 지원을 위한 모듈로 1.4 버전 이상에서 사용 가능하다. (pip show sqlalchemy)

데이터베이스 모델 정의
- aiomysql은 python의 비동기 I/O 지원을 위한 asyncio와 mysql 사이의 비동기적인 어댑터이다.
- FastAPI 어플리케이션에서 비동기 SQLAlchemy를 사용하고 MySQL 데이터베이스와 비동기적으로 통신하기 위해 필요하다.
  pip install aiomysql==0.2.0
  pip install sqlalchemy==2.0.27
---
from fastapi import FastAPI Depends
from sqlalchemy import Column, Integer, String, MetaData
from sqlalchemy.ext.asyncio import create_async_engine, AsyncSession
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import sessionmaker
from pydantic import BaseModel
from sqlalchemy.future import select
from contextlib import asynccontextmanager

# 비동기 데이터베이스 설정을 위한 문자열 정의 (사용자 이름, 비밀번호, 서버주소, 데이터베이스 이름)
DATABASE_URL = "mysql+aiomysql://pfeelone:pfeelone@localhost/db_name"

# 비동기 엔진 생성 (echo 옵션은 디버깅을 위한 것으로 모든 SQL 쿼리가 콘솔이나 로그에 출력된다.)
engine = create_async_engine(DATABASE_URL, echo=True)

# 비동기 세션 생성을 위한 세션메이커 정의
AsyncSessionLocal = sessionmaker(
    autocommit=False, autoflush=False, bind=engine, class_=AsyncSession
)

# SQLAlchemy 모델 기본 클래스 선언
Base = declarative_base()

class User(Base):
    __tablename__ = 'users'
    id = Column(Integer, primary_key=True, index=True)
    username = Column(String(50), unique=True, index=True)
    email = Column(String(120))
    
async def get_db():
    async with AsyncSessionLocal() as session:
        yield session # 비동기 제너레이터 패턴을 사용하여 세션 객체를 호출자에게 제공한다. FastAPI의 의존성 주입 시스템과 함께 사용되어 요청 처리 동안 해당 세션을 사용할 수 있게 해준다.
        await session.commit() # 세션의 변경사항을 데이터베이스에 비동기적으로 커밋한다.
   
# 비동기 컨텍스트 매니저를 쉽게 생성할 수 있게 해주는 데코레이터
@asynccontextmanager
async def app_lifespan(app: FastAPI):
    # engine.begin()은 SQLAlchemy의 비동기 엔진에서 트랜잭션을 시작하는 비동기 컨텍스트 매니저로 데이터베이스 작업을 트랜잭션 단위로 관리할 떄 사용된다.
    async with engine.begin() as conn:
        await conn.run_sync(Base.metadata.create_all)
    yield
    
app = FastAPI(lifespan=app_lifespan)
---
"""
```

```python
"""
HTTPBasic
- 실제로 많이 쓰이지 않지만 가장 기본적인 인증 메커니즘이다.

인증을 위한 HTTP 헤더
- HTTP 헤더는 클라이언트와 서버 간 데이터 교환 시 중요한 역할을 한다.
- Authorization 헤더는 인증 정보를 담는다.
  ex) Authorization:Basic {인코딩된 문자열} # 아이디:비밀번호를 Base64로 인코딩한 값
  
Base64
- 바이너리 데이터를 텍스트 형식으로 인코딩하는 방법으로 이메일을 통해 바이너리 데이터를 전송하거나 데이터를
  URL에 포함시킬 때 사용된다. (이메일은 텍스트만 지원하므로)
- 인코딩 과정은 다음과 같다.
 1. 바이너리 데이터를 6비트 단위로 나눈다.
 2. 각 6비트 값을 Base64 문자 집합에 매핑한다.
 3. 인코딩된 문자열의 길이가 4의 배수가 되도록 = 기호로 패딩한다.
- Base64로 인코딩된 데이터는 원본 데이터 크기의 약 33% 정도 증가하나 텍스트 기반 프로토콜에서 안전하게 전송할
  수 있다는 장점이 있다. 그러나 암호화 방식이 아니므로 민감한 정보를 보호하기에는 적합하지 않다.
  
FastAPI에서의 기본 인증 구현
---
from fastapi import FastAPI, Depends, HTTPException
from fastapi.security import HTTPBasic, HTTPBasicCredentials

app = FastAPI()

# HTTP Basic 인증 처리를 위해 필요한 보안 객체 생성
security = HTTPBasic()

# 인증 검사 함수 정의
# security에는 아이디와 패스워드가 Base64로 인코딩된 채로 저장되어 있다.
# HTTPBasicCredentials는 속성 값으로 아이디와 패스워드 변수를 가지고 있다. (디코딩이 된 채로 가져옴)
def get_current_username(credentials: HTTPBasicCredentials = Depends(security)):
    if credentials.username != "alice" or credentials.password != "password":
        raise HTTPException(status_code=401, detail="Unauthorized")
    return credentials.username
    
@app.get("/users/me")
def read_current_user(username: str = Depends(get_current_username)):
    return {"username": username}
---
"""
```

```python
"""
쿠키와 세션
- HTTP는 Sateless로 통신이 끝나면 상태를 유지하지 않으며 이를 보완하기 위한 기법이
  쿠키와 세션이다.

쿠키
- 클라이언트에서 서버에 요청 시 서버의 응답 헤더에 Set-Cookie 정보를 넣어 보낸다.
- 이후 클라이언트는 해당 정보를 통해 연결 상태를 유지할 수 있다. (ID, PW 정보)
- ID, PW에 대한 보안 이슈가 있다.

세션(Session)
- 웹에서 상태를 유지하는 중요한 방법
- 첫 요청 시 서버는 클라이언트에 고유한(암호화된) 세션 ID를 발급한다. (Set-cookie)
- 세션 ID는 클라이언트 측 쿠키에 저장된다.
- 클라이언트의 재요청 시 쿠키에 저장된 세션 ID를 서버에 전송한다.
- 서버는 세션 ID를 통해 클라이언트를 식별하고 정보를 저장하거나 불러온다.

SessionMiddleware
- FastAPI는 Starlette의 기능을 상속받아 세션 관리 기능을 제공한다.
- SessionMiddleware는 Starlette의 미들웨어로 세션 관리를 수행한다.
---
from fastapi import FastAPI, Request
from starlette.middleware.sessions import SessionMiddleware

app = FastAPI()
app.add_middleware(SessionMiddleware, secret_key="your-secret-key")

# 세션 설정
@app.post("/set/")
async def set_session(request: Request):
    request.session["username"] = "john"
    return {"message": "Session value set"}
    
# 세션 조회
@app.get("/get/")
async def get_session(request: Request):
    # Guest는 username이 없을 경우의 디폴트 값
    username = request.session.get("username", "Guest")
    return {"username": username}
---

curl을 사용한 테스트 방법
---
# 쿠키를 cookies.txt 파일에 저장
curl -X 'POST' 'http://127.0.0.1:8000/set/' -i -c cookies.txt

# 세션 값 가져오기 (cookies 파일에서 쿠키를 읽어 요청에 포함)
curl -X 'GET' 'http://127.0.0.1:8000/get/' -i -b cookies.txt
---
- 쿠키 파일 내용
 : 첫 번째 FALSE는 쿠키가 전송되어야 하는 경우에 대한 보안 연결(HTTPS)의 필요성을
   나타낸다. (FALSE이면 HTTP로도 전송될 수 있다.)
 : 두 번째 FALSE는 쿠키의 Secure 속성과 관련이 있으며 True일 경우 쿠키는 오직
   HTTPS 연결을 통해서만 전송된다.

"""
```
